# Introduction


Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.__version__
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(4)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
plt.scatter(x="total_bill", y="tip", data=tips);
```

```python
sns.set(style="darkgrid")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
tips.time.value_counts()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="time",
            data=tips);
```

```python
plt.figure(figsize=(12,12))
sns.scatterplot(x="total_bill", y="tip",size="size", sizes = (2,200), data=tips)
```

```python
#facetting
sns.relplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips);
```

```python
np.random.seed(10)
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df,height=8)
```

# Categorical Variables

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

```python
sns.boxplot(x="day", y="total_bill", data=tips);
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day",data=tips,kind="count")
```

```python
sns.catplot(x="day",data=tips,kind="count",col="smoker")
```

```python
sns.catplot(x="size",data=tips,kind="count",col="smoker",row="sex")
```

```python
sns.catplot(y="tip",x="day",data=tips,kind="bar",hue="sex",ci=None)
```

# Distribution

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

```python
sns.distplot(x,hist=False);
```

```python
sns.distplot(x,kde=False,color="red");
```

```python
sns.kdeplot(tips.total_bill,shade=True)
```
